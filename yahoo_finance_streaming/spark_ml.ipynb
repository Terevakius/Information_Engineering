```python
import numpy as np

# Create training and testing set

def create_test_training_data(df_predict):
    # Get the number of rows in the Spark DataFrame
    row_count = df_predict.count()
    # Divide the DataFrame into the training data
    training_data = df_predict.limit(row_count - 1)
    # Get the last row as test data
    test_data = df_predict.subtract(training_data)
    
    return training_data, test_data


def create_label_for_lr(df):
    # Compare the 'open' column with the shifted 'previous_open' column
    df['next_open_larger_than_previous_open'] = (df['open'] < df['open'].shift(-1)).astype(int)
    # Fill the first row of the new column with 0
    df['next_open_larger_than_previous_open'].fillna(0, inplace=True)
    
    return df


def load_file(input_file):
    df = pd.read_csv(input_file)
    
    return df
```

```python
import pandas as pd
import numpy as np
import os
import time

import pyspark
import pyspark.sql
import sparky
import slash
sc = sparky.connect("ml1-logreg-...", 2)
spark = pyspark.sql.SparkSession.builder.getOrCreate()

from pyspark.ml import Pipeline
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.classification import LogisticRegression
from pyspark.sql.types import StructType, StructField, DoubleType, IntegerType

import matplotlib.pyplot as plt
import mplfinance as mpf


stock_directory = os.path.join(path, ticker)
# Get the initial list of files in the folder
initial_files = os.listdir(stock_directory)

return_window_cum = 0

while True:
    # Delay the loop for e.g. 1 second
    time.sleep(1)
    
    # Get the current list of files in the folder
    current_files = os.listdir(stock_directory)
    #print('Current file : \n')
    #print(current_files)
    
    # Check if any new files are added
    new_files = [file for file in current_files if file not in initial_files]
    
    if new_files:
        #print("New files added, namely: \n")
        for file in new_files:
            input_path = os.path.join(stock_directory, file)
            df = load_file(input_path)
            #print('This is df: \n')
            #print(df.head())
        
            # Update the initial file list for the next iteration
            initial_files = current_files
            
            # Create New Label
            df = create_label_for_lr(df)
            # Keep only relevant variables for prediction
            df_predict = df[['next_open_larger_than_previous_open', 'open', 'close', 'MA5', 'MA15']]

            # Rename first column to "label"
            df_predict = df_predict.rename(columns={'next_open_larger_than_previous_open': 'label'})
            # Remove rows with missing values from df
            df_predict = df_predict.dropna()
            #print('This is df_predict head')
            #print(df_predict.head())
            
            # Define the schema for the training DataFrame
            schema = StructType([
                StructField('label', IntegerType()),
                StructField('open', DoubleType()),
                StructField('close', DoubleType()),
                StructField('MA5', DoubleType()),
                StructField('MA15', DoubleType())
            ])

            # Convert pandas DataFrame to Spark DataFrame
            spark_df = spark.createDataFrame(df_predict, schema=schema)
            # You can perform Spark operations on the spark_df

            # Split into training and test data
            training, testing = create_test_training_data(spark_df)
            #training.show()
            
            # Configure an ML pipeline, which consists of two stages: feature assembler and lr.
            # Transform n feature vectors into one single vector column
            assembler = VectorAssembler(inputCols=training.columns[1:], outputCol='features')
            #print('this is the assember: \n')
            #print(assembler)
            # check how the new column "features" look like
            assembler.transform(training).head()
            lr = LogisticRegression(maxIter=10, regParam=0.01)
            pipeline = Pipeline(stages=[assembler, lr])
            #print("LogisticRegression parameters:\n" + lr.explainParams() + "\n")

            # predict 
            #print('this is training')
            #training.show()
            model = pipeline.fit(training)
            prediction = model.transform(testing)
            #print('This is the testing: \n')
            #testing.show()
            #print('This is the prediction: \n')
            #prediction.show()
                        
            testing_pandas = testing.toPandas()
            prediction_pandas = prediction.toPandas()
            
            if prediction_pandas['prediction'][0] == 0:
                print('Short sell signal generated! \n')
                return_window = (testing_pandas['open'][0] / testing_pandas['close'][0]) - 1
            elif prediction_pandas['prediction'][0] == 1:
                print("Buy signal generated based on Logistic Regression! \n")
                return_window = (testing_pandas['close'][0] / testing_pandas['open'][0]) - 1
            
            if return_window > 0:
                print("Lucky you - the prediction was very good. The return for that window is: " + str("{:.5%}".format(return_window)) + ". Congrats! \n")
            else:
                print("Bad luck - the prediction was wrong. The return for that window is: " + str("{:.5%}".format(return_window)) + ". \n")
            
            return_window_cum = return_window_cum + return_window
            print("The cumulative return over observed windows is: " + str("{:.2%}".format(return_window_cum)) + "! \n")
                
            # Create Candlestick Plot
            candlestick_df = df[['time_stamp', 'open', 'high', 'low', 'close']]
            candlestick_df['time_stamp'] = pd.to_datetime(candlestick_df['time_stamp'])
            candlestick_df.set_index('time_stamp', inplace=True)

            # Plot the candlestick chart
            mpf.plot(candlestick_df, type='candle', style='yahoo')
            plt.show()

            
```
